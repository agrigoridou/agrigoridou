# TASK 2

Φορτώνει τα δεδομένα από ένα αρχείο (π.χ. test.txt).
Χρησιμοποιεί την συνάρτηση load_sentences() για να διαβάσει και να επεξεργαστεί τις προτάσεις, χωρίζοντας κάθε λέξη σε tokens, POS tags, chunk tags και NER tags.

```python
import torch
import random

# Αρχική διαδικασία φόρτωσης των δεδομένων
base_path = '/kaggle/input/conll003-englishversion/'  # Αντικατάστησε με το σωστό path για τα δεδομένα σου

# Η συνάρτηση φόρτωσης των προτάσεων
def load_sentences(filepath):
    sentences = []
    tokens = []
    pos_tags = []
    chunk_tags = []
    ner_tags = []
    with open(filepath, 'r') as f:
        for line in f.readlines():
            if (line == ('-DOCSTART- -X- -X- O\n') or line == '\n'):
                if len(tokens) > 0:
                    sentences.append({'tokens': tokens, 'pos_tags': pos_tags, 'chunk_tags': chunk_tags, 'ner_tags': ner_tags})
                    tokens = []
                    pos_tags = []
                    chunk_tags = []
                    ner_tags = []
            else:
                l = line.split(' ')
                tokens.append(l[0])
                pos_tags.append(l[1])
                chunk_tags.append(l[2])
                ner_tags.append(l[3].strip('\n'))
    return sentences

```

# Επιλογή τυχαίας πρότασης από το test set

Φόρτωση του Test Set: Ο κώδικας ξεκινά με την φόρτωση των δεδομένων από το αρχείο test.txt μέσω της συνάρτησης load_sentences(). Αυτή η συνάρτηση δημιουργεί μια λίστα από προτάσεις, όπου κάθε πρόταση είναι ένα λεξικό που περιλαμβάνει τα tokens, τις pos_tags, τις chunk_tags και τις ner_tags.

Τυχαία Επιλογή Πρότασης: Επιλέγεται τυχαία μια πρόταση από το test set που περιέχει τουλάχιστον 10 tokens, χρησιμοποιώντας την random.choice() σε έναν φιλτραρισμένο κατάλογο από τις προτάσεις που πληρούν αυτό το κριτήριο.

Εκτύπωση Προτάσεων και NER Tags: Η τυχαία επιλεγμένη πρόταση εκτυπώνεται μαζί με τα αντίστοιχα NER tags της. Τα tokens είναι οι λέξεις που αποτελούν την πρόταση και τα ner_tags είναι οι ετικέτες που αναγνωρίζει το μοντέλο για κάθε λέξη.

```python
import random

# Φόρτωση του test set
test_sentences = load_sentences(base_path + 'test.txt')

# Εύρεση τυχαίας πρότασης με τουλάχιστον 10 tokens από το test set
random_sentence = random.choice([s for s in test_sentences if len(s['tokens']) >= 10])

# Εκτύπωση της τυχαίας πρότασης και των αντίστοιχων NER tags
print("Random sentence:")
print("Tokens:", random_sentence['tokens'])
print("NER Tags:", random_sentence['ner_tags'])
```

Σε αυτή την τυχαία πρόταση, οι περισσότερες ετικέτες είναι O (που σημαίνει ότι το token δεν είναι μέρος κάποιας ονοματικής οντότητας). Η μόνη εξαίρεση είναι η λέξη Scottish, η οποία έχει την ετικέτα B-MISC (αναγνωρίζεται ως ονοματική οντότητα τύπου Miscellaneous).

# Εύρεση σωστών και λάθος προβλέψεων

Αυτός ο κώδικας συγκρίνει τις πραγματικές ετικέτες της τυχαίας πρότασης με τις προβλέψεις του μοντέλου και τις διαχωρίζει σε σωστές και λάθος προβλέψεις.

```python
# Αναγνώριση των σωστών και λανθασμένων ετικετών NER
tokens = random_sentence['tokens']
ner_tags = random_sentence['ner_tags']

# Ανάλυση των σωστών και λανθασμένων προβλέψεων
correct_tags = []
incorrect_tags = []

for token, tag in zip(tokens, ner_tags):
    if tag == "O":
        incorrect_tags.append((token, tag))
    else:
        correct_tags.append((token, tag))

print("Correct Tags:")
for token, tag in correct_tags:
    print(f"Token: {token}, Tag: {tag}")

print("\nIncorrect Tags:")
for token, tag in incorrect_tags:
    print(f"Token: {token}, Tag: {tag}")
```

Η μόνη σωστή πρόβλεψη είναι για τη λέξη Scottish, η οποία αναγνωρίζεται σωστά ως ονοματική οντότητα τύπου B-MISC. Όλες οι υπόλοιπες λέξεις έχουν ετικέτα O, που σημαίνει ότι δεν αναγνωρίζονται ως ονοματικές οντότητες.

# Εισαγωγή νέας πρότασης από online πηγή

Το μοντέλο BERT φορτώνεται από την προεκπαιδευμένη έκδοση του dbmdz/bert-large-cased-finetuned-conll03-english, η οποία έχει εκπαιδευτεί για NER στην αγγλική γλώσσα.

```python
from transformers import BertForTokenClassification, BertTokenizer
import torch

# Φορτώνουμε το μοντέλο και τον tokenizer
model = BertForTokenClassification.from_pretrained('dbmdz/bert-large-cased-finetuned-conll03-english')
tokenizer = BertTokenizer.from_pretrained('dbmdz/bert-large-cased-finetuned-conll03-english')

```

Αυτό το κομμάτι του κώδικα φορτώνει τη νέα πρόταση και τη στέλνει στο μοντέλο BERT για να προβλέψει τις ετικέτες NER για κάθε token.

```python
# Δημιουργία του tagmap
tagmap = {0: 'O', 1: 'B-PER', 2: 'I-PER', 3: 'B-ORG', 4: 'I-ORG', 5: 'B-LOC', 6: 'I-LOC', 7: 'B-MISC', 8: 'I-MISC'}

# Εισαγωγή νέας πρότασης
new_sentence = "Tesla has reported a significant rise in their stock price after the announcement of new electric vehicle models."

# Tokenization της νέας πρότασης
inputs = tokenizer(new_sentence.split(), return_tensors="pt", padding=True, truncation=True)

# Προβλέψεις με το μοντέλο
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    predictions = torch.argmax(outputs.logits, dim=2)

# Αποθήκευση των tokens και των NER tags
tokens = new_sentence.split()
predicted_tags = [tagmap[idx.item()] for idx in predictions[0]]

# Εκτύπωση της νέας πρότασης και των αντίστοιχων προβλέψεων
print("New sentence:")
print("Tokens:", tokens)
print("Predicted NER Tags:", predicted_tags)
```

Το μοντέλο προβλέπει τις ετικέτες NER για τα tokens της νέας πρότασης. Στη συγκεκριμένη περίπτωση, η λέξη Tesla ανιχνεύεται ως οργανισμός (I-ORG) και η λέξη has ανήκει στην ετικέτα προσωπικής οντότητας (I-PER). Οι υπόλοιπες λέξεις δεν ανιχνεύονται ως ονοματικές οντότητες και λαμβάνουν την ετικέτα O.

# Ανάλυση των σωστών και λάθος προβλέψεων στην νέα πρόταση

```python
# Αναγνώριση των σωστών και λανθασμένων προβλέψεων
correct_tags = []
incorrect_tags = []

for token, tag in zip(tokens, predicted_tags):
    if tag != "O":  # Αν το tag είναι διαφορετικό από "O", σημαίνει ότι έγινε πρόβλεψη για NER
        correct_tags.append((token, tag))  # Αυτή η πρόβλεψη θεωρείται σωστή
    else:
        incorrect_tags.append((token, tag))  # Αν είναι "O", θεωρείται λανθασμένη πρόβλεψη

print("Correct Tags for the new sentence:")
for token, tag in correct_tags:
    print(f"Token: {token}, Tag: {tag}")

print("\nIncorrect Tags for the new sentence:")
for token, tag in incorrect_tags:
    print(f"Token: {token}, Tag: {tag}")
```

Σωστές Προβλέψεις:

Η λέξη has αναγνωρίζεται σωστά ως μέρος ενός οργανισμού (I-ORG).
Η λέξη reported αναγνωρίζεται σωστά ως προσωπική οντότητα (I-PER).

Λανθασμένες Προβλέψεις:

Η λέξη Tesla που είναι το όνομα της εταιρείας, εκ παραδρομής δεν αναγνωρίζεται ως οργανισμός, με την ετικέτα O.
Όλες οι υπόλοιπες λέξεις (a, significant, rise, κ.λπ.) δεν αναγνωρίζονται ως ονοματικές οντότητες και κατηγοριοποιούνται ως O, που σημαίνει "όχι ονοματική οντότητα".

## Ρόλος της Συνάρτησης align_label:

Η συνάρτηση align_label έχει ως σκοπό να ευθυγραμμίσει τις ετικέτες (labels) με τα tokens που εισάγονται στο μοντέλο. Αυτό είναι ιδιαίτερα χρήσιμο όταν χρησιμοποιούμε μια διαδικασία tokenization για να χωρίσουμε τις προτάσεις σε μικρότερα κομμάτια (tokens). Επειδή η διαδικασία tokenization μπορεί να κατατμήσει τα αρχικά tokens σε υπο-λέξεις (subwords), οι ετικέτες (labels) των αρχικών tokens μπορεί να μην ταιριάζουν πλέον απευθείας με τα υπο-λέξεις.

Για παράδειγμα, ας πούμε ότι έχουμε το token "Tesla", και το tokenizer το χωρίζει σε δύο υπο-λέξεις, όπως ["Tes", "la"]. Η ετικέτα για το αρχικό token "Tesla" (π.χ., "B-ORG" για οργανισμούς) πρέπει να ανατεθεί σε όλα τα υπο-λέξεις που παράγονται από αυτό το token.

## Σκοπός του -100:

Το -100 χρησιμοποιείται στην align_label για να δείξει ότι τα υπο-λέξεις που προκύπτουν από το tokenization δεν πρέπει να ληφθούν υπόψη κατά την εκπαίδευση. Αυτό συμβαίνει επειδή η ετικέτα για αυτά τα υπο-λέξεις δεν έχει νόημα και δεν πρέπει να επηρεάσει την εκπαίδευση του μοντέλου.

```python
import torch
import random

# Αρχική διαδικασία φόρτωσης των δεδομένων
base_path = '/kaggle/input/conll003-englishversion/'  # Αντικατάστησε με το σωστό path για τα δεδομένα σου

# Η συνάρτηση φόρτωσης των προτάσεων
def load_sentences(filepath):
    sentences = []
    tokens = []
    pos_tags = []
    chunk_tags = []
    ner_tags = []
    with open(filepath, 'r') as f:
        for line in f.readlines():
            if (line == ('-DOCSTART- -X- -X- O\n') or line == '\n'):
                if len(tokens) > 0:
                    sentences.append({'tokens': tokens, 'pos_tags': pos_tags, 'chunk_tags': chunk_tags, 'ner_tags': ner_tags})
                    tokens = []
                    pos_tags = []
                    chunk_tags = []
                    ner_tags = []
            else:
                l = line.split(' ')
                tokens.append(l[0])
                pos_tags.append(l[1])
                chunk_tags.append(l[2])
                ner_tags.append(l[3].strip('\n'))
    return sentences
```

## Επιλογή τυχαίας πρότασης από το test set

Φόρτωση του Test Set: Ο κώδικας ξεκινά με την φόρτωση των δεδομένων από το αρχείο test.txt μέσω της συνάρτησης load_sentences(). Αυτή η συνάρτηση δημιουργεί μια λίστα από προτάσεις, όπου κάθε πρόταση είναι ένα λεξικό που περιλαμβάνει τα tokens, τις pos_tags, τις chunk_tags και τις ner_tags.

Τυχαία Επιλογή Πρότασης: Επιλέγεται τυχαία μια πρόταση από το test set που περιέχει τουλάχιστον 10 tokens, χρησιμοποιώντας την random.choice() σε έναν φιλτραρισμένο κατάλογο από τις προτάσεις που πληρούν αυτό το κριτήριο.

Εκτύπωση Προτάσεων και NER Tags: Η τυχαία επιλεγμένη πρόταση εκτυπώνεται μαζί με τα αντίστοιχα NER tags της. Τα tokens είναι οι λέξεις που αποτελούν την πρόταση και τα ner_tags είναι οι ετικέτες που αναγνωρίζει το μοντέλο για κάθε λέξη.

```python
import random

# Φόρτωση του test set
test_sentences = load_sentences(base_path + 'test.txt')

# Εύρεση τυχαίας πρότασης με τουλάχιστον 10 tokens από το test set
random_sentence = random.choice([s for s in test_sentences if len(s['tokens']) >= 10])

# Εκτύπωση της τυχαίας πρότασης και των αντίστοιχων NER tags
print("Random sentence:")
print("Tokens:", random_sentence['tokens'])
print("NER Tags:", random_sentence['ner_tags'])
```

## Εύρεση σωστών και λάθος προβλέψεων

Αυτός ο κώδικας συγκρίνει τις πραγματικές ετικέτες της τυχαίας πρότασης με τις προβλέψεις του μοντέλου και τις διαχωρίζει σε σωστές και λάθος προβλέψεις.

```python
# Αναγνώριση των σωστών και λανθασμένων ετικετών NER
tokens = random_sentence['tokens']
ner_tags = random_sentence['ner_tags']

# Ανάλυση των σωστών και λανθασμένων προβλέψεων
correct_tags = []
incorrect_tags = []

for token, tag in zip(tokens, ner_tags):
    if tag == "O":
        incorrect_tags.append((token, tag))
    else:
        correct_tags.append((token, tag))

print("Correct Tags:")
for token, tag in correct_tags:
    print(f"Token: {token}, Tag: {tag}")

print("\nIncorrect Tags:")
for token, tag in incorrect_tags:
    print(f"Token: {token}, Tag: {tag}")
```

Τα αποτελέσματα δείχνουν ότι το μοντέλο αναγνωρίζει σωστά οντότητες όπως εθνικότητες ("Polish") και οργανισμούς ("Zywiec"), αλλά αποτυγχάνει να εντοπίσει χρονολογίες ("1996"), ποσοστά ("10-percent") και επαγγελματικούς ρόλους ("chief executive"). Η υπερβολική επισήμανση των tokens ως O (Outside) υποδηλώνει περιορισμένη ικανότητα διάκρισης σημαντικών από μη σημαντικά tokens. Το μοντέλο χρειάζεται βελτίωση μέσω εκπαίδευσης με περισσότερα παραδείγματα για αριθμητικά δεδομένα, χρονολογίες και επαγγελματικούς τίτλους, καθώς και χρήση πιο ισχυρών pre-trained μοντέλων ή κανόνων για post-processing. Συνολικά, αναγνωρίζει βασικές οντότητες, αλλά έχει αδυναμίες στις πιο σύνθετες περιπτώσεις.

## Εισαγωγή νέας πρότασης από online πηγή

Το μοντέλο BERT φορτώνεται από την προεκπαιδευμένη έκδοση του dbmdz/bert-large-cased-finetuned-conll03-english, η οποία έχει εκπαιδευτεί για NER στην αγγλική γλώσσα.

```python
from transformers import BertForTokenClassification, BertTokenizer
import torch

# Φορτώνουμε το μοντέλο και τον tokenizer
model = BertForTokenClassification.from_pretrained('dbmdz/bert-large-cased-finetuned-conll03-english')
tokenizer = BertTokenizer.from_pretrained('dbmdz/bert-large-cased-finetuned-conll03-english')
```

```python
# Δημιουργία του tagmap
tagmap = {0: 'O', 1: 'B-PER', 2: 'I-PER', 3: 'B-ORG', 4: 'I-ORG', 5: 'B-LOC', 6: 'I-LOC', 7: 'B-MISC', 8: 'I-MISC'}

# Εισαγωγή νέας πρότασης
new_sentence = "Tesla has reported a significant rise in their stock price after the announcement of new electric vehicle models."

# Tokenization της νέας πρότασης
inputs = tokenizer(new_sentence.split(), return_tensors="pt", padding=True, truncation=True)

# Προβλέψεις με το μοντέλο
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    predictions = torch.argmax(outputs.logits, dim=2)

# Αποθήκευση των tokens και των NER tags
tokens = new_sentence.split()
predicted_tags = [tagmap[idx.item()] for idx in predictions[0]]

# Εκτύπωση της νέας πρότασης και των αντίστοιχων προβλέψεων
print("New sentence:")
print("Tokens:", tokens)
print("Predicted NER Tags:", predicted_tags)
```

## Ανάλυση των σωστών και λάθος προβλέψεων στην νέα πρόταση

```python
# Αναγνώριση των σωστών και λανθασμένων προβλέψεων
correct_tags = []
incorrect_tags = []

for token, tag in zip(tokens, predicted_tags):
    if tag != "O":  # Αν το tag είναι διαφορετικό από "O", σημαίνει ότι έγινε πρόβλεψη για NER
        correct_tags.append((token, tag))  # Αυτή η πρόβλεψη θεωρείται σωστή
    else:
        incorrect_tags.append((token, tag))  # Αν είναι "O", θεωρείται λανθασμένη πρόβλεψη

print("Correct Tags for the new sentence:")
for token, tag in correct_tags:
    print(f"Token: {token}, Tag: {tag}")

print("\nIncorrect Tags for the new sentence:")
for token, tag in incorrect_tags:
    print(f"Token: {token}, Tag: {tag}")
```

Τα αποτελέσματα δείχνουν σημαντικά σφάλματα στις προβλέψεις του μοντέλου για τη νέα πρόταση. Συγκεκριμένα, το μοντέλο απέτυχε να αναγνωρίσει το "Tesla" ως οργανισμό (B-ORG), μια προφανή οντότητα που θα έπρεπε να εντοπιστεί. Αντίθετα, λανθασμένα επισήμανε τα "has" και "reported" ως I-ORG και I-PER αντίστοιχα, κατηγοριοποιώντας λέξεις που δεν είναι οντότητες. Αυτό υποδηλώνει ότι το μοντέλο παρουσιάζει περιορισμένη ικανότητα διάκρισης μεταξύ κανονικών λέξεων και οντοτήτων, καθώς και λανθασμένη ερμηνεία της σύνταξης της πρότασης.
